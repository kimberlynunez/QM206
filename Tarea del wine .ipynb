```python
wine = datasets.load_wine()
wine_dataframe=pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_dataframe['target']=wine.target
x=wine_dataframe['total_phenols']
print(x.mean())
y=wine_dataframe['proline']
col=(wine_dataframe['target']+1)*2
print(y.mean())
total_phenol=wine_dataframe['total_phenols']*10
proline=wine_dataframe['proline']*10
```

```python

plt.scatter(x,y,c=col,s=60,alpha=0.5)
plt.xlabel('Fenoles tpor elotales (%)')
plt.ylabel('Prolina (%)')
plt.title('Porcentaje de Prolina vs Porcentaje de Fenoles')
plt.grid(True)
plt.show()
```

```python
##Descripcion del grafico 

wine = datasets.load_wine()
wine_dataframe=pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_dataframe['target']=wine.target
x=wine_dataframe['color_intensity']
print(x.mean())
y=wine_dataframe['ash']
col=(wine_dataframe['target']+1)*2
print(y.mean())
CI=wine_dataframe['color_intensity']*10
ASH=wine_dataframe['ash']*10
```
