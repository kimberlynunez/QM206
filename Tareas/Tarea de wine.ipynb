```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine=datasets.load_wine()
wine
```

```python
wine_dataframe=pd.DataFrame(data=wine.data,columns=wine.feature_names)
wine_dataframe
 
```

## Base de datos Wine

Estos datos son el resultado de un análisis químico de vinos cultivados en la misma región en Italia pero derivados de tres cultivares diferentes. El análisis determinó las cantidades de 13 componentes encontrados en cada uno de los tres tipos de vinos. 

Creo que el conjunto de datos inicial tenía alrededor de 30 variables, pero por alguna razón solo tengo la versión de 13 dimensiones. Tenía una lista de cuáles eran las 30 o más variables, pero a.) Lo perdí, y b), no sabría qué 13 variables están incluidas en el conjunto. 

Los atributos son (donados por Riccardo Leardi, riclea '@' anchem.unige.it ) 
1) Alcohol 
2) Ácido málico 
3) Cenizas 
4) Alcalinidad de cenizas 
5) Magnesio 
6) Fenoles totales 
7) Flavonoides 
8) Fenoles no flavonoides 
9) Proantocianinas 
10) Intensidad del color 
11) Hue 
12) OD280 / OD315 de los vinos diluidos 
13) Prolina 


```python
wine_dataframe["target"]=wine.target
wine_dataframe
x=wine_dataframe["alcohol"]
print(x.mean())
y=wine_dataframe["malic_acid"]
col=(wine_dataframe["target"]+1)*2
print(y.mean())
alcohol=wine_dataframe["alcohol"]*50
malic_acid=wine_dataframe["malic_acid"]



```

```python
plt.scatter(x,y,c=col,s=alcohol,alpha=0.5)
plt.xlabel("alcohol")
plt.ylabel("malic_acid")
plt.title("alcohol vs malic_acid")
plt.grid(True)
plt.show()
```
