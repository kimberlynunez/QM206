```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x=np.arange(10)
y=np.sin(x)
plt.plot(x,y,"ro-")
plt.xlabel("distancia(m)")
plt.ylabel("densidad(g/mL)")
plt.show()
```
