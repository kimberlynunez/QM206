```python
import matplotlib.pyplot as plt
import numpy as np 
```

## cinética de primer orden 


```python
A=1.0
B=0
k1=0.1
dt=0.1
t=0
x=[]
yA=[]
yB=[]
while t<100:
    x.append(t)
    yA.append(A)
    yB.append(B)
    dA=-k1*A*dt
    dB=+k1*A*dt
    A+=dA
    B+=dB
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.xlabel("t(m)")
plt.ylabel("Concentracion")
plt.title("Cinetica de primer orden")
plt.show()
```

## Segunda reacccion cinetica consecutiva


A=1.0

k1=0.1

    x.append(t)

    A+=dA

    C+=dC
    t+=dt

## Equilibrio químico

```python
A=1.0
B=0
k1=0.1
k2=0.2
dt=0.1
t=0
x=[]
yA=[]
yB=[]
while t<100:
    x.append(t)
    yA.append(A)
    yB.append(B)
    dA=-k1*A*dt+k2*B*dt
    dB=+k1*A*dt-k2*B*dt
    A+=dA
    B+=dB
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.xlabel("t(m)")
plt.ylabel("Cinetica de primer orden")
plt.show()
```
