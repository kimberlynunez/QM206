```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deltaS_N2=216.866
deltaH_N2=0*1000
deltaS_H2=155.606
deltaH_H2=0*1000
deltaS_NH3=227.829
deltaH_NH3=-52.618*1000



T=700    #T en Kelvin
pt=200      #presion en bar
R=8.314     #J/K-mol
```

```python
uN2_est=deltaH_N2-T*deltaS_N2
uH2_est=deltaH_H2-T*deltaS_H2
uNH3_est=deltaH_NH3-T*deltaS_NH3
print(uN2_est,uH2_est,uNH3_est)
```

```python
x=np.arange(0.001,1,0.001)
n_N2=x
n_H2=3*x
n_NH3=2-2*x
nt=2+2*x


X_N2=n_N2/nt
X_H2=n_H2/nt
X_NH3=n_NH3/nt
```

```python
uN2=uN2_est+R*T*np.log(X_N2)
uH2=uH2_est+R*T*np.log(X_H2)
uNH3=uNH3_est+R*T*np.log(X_NH3)
uT=n_N2*uN2+n_H2*uH2+n_NH3*uNH3

minimo=np.min(uT)
n=np.where(minimo==uT)
xeq=x[n]
print(xeq)

plt.plot(x,uT)
plt.xlabel(r'$\xi(mol)$')
plt.ylabel(r'$\Delta G(J/mol)$')
plt.axvline(x=xeq,color="green",linestyle="-.")
plt.text(xeq,-78000,r'$\xi_{eq}=$'+str(xeq),fontsize=14)
plt.show()
```
