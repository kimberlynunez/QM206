```python
from sklearn import datasets 
import pandas as pd
import matplotlib.pyplot as plt 
 
```

```python
iris=datasets.load_iris()
iris_dataframe=pd.DataFrame(data=iris.data,columns=["A","B","C","D"])
iris_dataframe["target"]=iris.target
x=iris_dataframe["A"]
print(x.mean())
y=iris_dataframe["B"]
col=(iris_dataframe["target"]+1)*2
print(y.mean())
C=iris_dataframe["C"]*50
```

```python
plt.scatter(x,y,c=col,s=C,alpha=0.5)
plt.xlabel("Largo del Spalo (cm)")
plt.ylabel("Ancho del Spalo (cm )")
plt.title("Ancho de Spalo vs Largo del Spalo ")
plt.grid(True)
plt.show()
```
