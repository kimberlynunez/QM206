```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A=1.0
B=2
C=0
k1=0.15
dt=0.1
t=0
x=[]
yA=[]
yB=[]
yC=[]
while t<100:
    x.append(t)
    yA.append(A)
    yB.append(B)
    yC.append(C)
    dA=-k1*A*dt
    dB=+k1*A*dt
    dC=+k1*A*dt
    A+=dA
    B+=dB
    C+=dC
    t+=dt
plt.plot(x,yA)
plt.plot(x,yB)
plt.plot(x,yC)
plt.xlabel("t(m)")
plt.ylabel("Concentracion")
plt.title("Cinetica de primer orden")
plt.show()
```
