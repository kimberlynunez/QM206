```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,1],[3,1,7,4],color="purple",linewidth=3,alpha=0.5,marker="o",markersize=10,markeredgewidth=5,markeredgecolor="black",markerfacecolor="blue")
plt.xlabel("$x(cm^2)$")
plt.ylabel("$Intensidad(\mu)$")
plt.title("Experimento No.1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.plot(x,y,color="pink")
plt.plot(x,y2,color="black",dashes=[2,2])
plt.title("funcion sin(4x) 0-$2\pi$",color="black")
plt.xlabel("t(seg)")
plt.ylabel("Amplitud de la onda ")
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*300+50
plt.scatter(x,y,s=r,color="red")
plt.show()
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=50,facecolor="purple",edgecolor="green",density=True)
plt.title("Distribucion de la poblacion ")
plt.ylabel("Densidad de habitantes/$m^2$")
plt.text(2, .35, r'$\mu=0,\ \sigma=15$')
plt.show()
```
